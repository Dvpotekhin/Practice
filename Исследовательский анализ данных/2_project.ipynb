Больше всего цена зависит от общей площади. Чем больше площадь, тем больше цена.

```python
def whatfloor_int(data):
    floor = data['what_floor']
    if floor == 'другой':
        return 0
    elif floor == 'первый':
        return 1
    else:
        return 2
```
